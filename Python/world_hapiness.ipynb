```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('/Users/DataForTable2.1WHR2023.xls')
```

```python
print(data.head())

print(data.describe())

print(data.isnull().sum())
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Life Ladder'], bins=20, kde=True)
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.title('Distribution of Happiness Score')
plt.show()
```

```python
correlation_matrix = data.corr()
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(data[['Life Ladder', 'Log GDP per capita', 'Healthy life expectancy at birth', 'Freedom to make life choices']], height=2)

plt.suptitle('Pairwise Relationships')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='Country name', palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Country Name')
plt.ylabel('Count')
plt.title('Distribution of Countries by Region')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='year', y='Life Ladder', errorbar=None)
plt.xlabel('year')
plt.ylabel('Life Ladder')
plt.title('Trend of Life Ladder Over the Years')
plt.show()
```

```python
country_name = 'Indonesia'
country_data = data[data['Country name'] == country_name]

plt.figure(figsize=(10, 6))
sns.lineplot(data=country_data, x='year', y='Life Ladder', marker='o')
plt.xlabel('Year')
plt.ylabel('Happiness Score')
plt.title(f'Trend of Happiness Score Over the Years for {country_name}')
plt.grid(True)
plt.show()
```

```python
countries = data[(data['Country name'].isin(['Indonesia','Japan', 'Taiwan Province of China', 'South Korea', 'Mozambique']))]
sns.set_style('whitegrid')
sns.set(font_scale=1.5)
sns.lineplot(x='year', y='Life Ladder', data=countries, hue='Country name',linewidth=4,palette= 'hls').set(title='Life Ladder by year at Indonesia, Japan, and France')

plt.show()
```

```python
countries = data[(data['Country name'].isin(['Russia', 'United States', 'France', 'Japan', 'India', 'Algeria']))]
sns.set_style('whitegrid')
sns.set(font_scale=1.5)
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(30,30))
sns.lineplot(x='year', y='Life Ladder', data=countries, hue='Country name',linewidth=4,palette= 'hls', ax=axes[0][0]).set(title='Life Ladder by year')
sns.lineplot(x='year', y='Log GDP per capita', data=countries, hue='Country name',linewidth=4,palette= 'hls', ax=axes[0][1]).set(title='GDP Impact every year')
sns.lineplot(x='year', y='Social support', data=countries, hue='Country name',linewidth=4,palette= 'hls', ax=axes[1][0]).set(title='Social Support by year')
sns.lineplot(x='year', y='Healthy life expectancy at birth', data=countries, hue='Country name',linewidth=4,palette= 'hls', ax=axes[1][1]).set(title='Healthy Life expectancy at birth')
sns.lineplot(x='year', y='Generosity', data=countries, hue='Country name',linewidth=4,palette= 'hls', ax=axes[2][0]).set(title='Generosity by year')
sns.lineplot(x='year', y='Freedom to make life choices', data=countries, hue='Country name',linewidth=4,palette= 'hls', ax=axes[2][1]).set(title='Freedom to people by year')
plt.show()
```
