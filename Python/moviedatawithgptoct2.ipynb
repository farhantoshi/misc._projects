```python
import requests
import pandas as pd
from IPython.display import IFrame
import openai
import json
```

```python
# Ask user to fill in the information
include_adult = input("Include adult content? (true/false): ")
include_video = input("Include videos? (true/false): ")
language = input("Language: ")
page = input("Page number: ")
sort_by = input("Sort by: ")
with_genres = input("Genres to include: ")
without_genres = input("Genres to exclude: ")

# Format the information into a URL query string
query_string = {
    "include_adult": include_adult,
    "include_video": include_video,
    "language": language,
    "page": page,
    "sort_by": sort_by,
    "with_genres": with_genres,
    "without_genres": without_genres
}

# Print the URL query string
print(f"URL query string: {query_string}")
```

```python
url = f"https://api.themoviedb.org/3/discover/movie?include_adult={query_string['include_adult']}&include_video={query_string['include_video']}&language={query_string['language']}&page={query_string['page']}&sort_by={query_string['sort_by']}&with_genres={query_string['with_genres']}&without_genres={query_string['without_genres']}"

headers = {
    "accept": "application/json",
    "Authorization": "Bearer AUTHORIZATION KEY"
}

response = requests.get(url, headers=headers)

print(response.text)
```

```python
# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()

    # Extract the list of movies from the response
    movies = data.get("results", [])

    # Create a DataFrame from the list of movies
    df = pd.DataFrame(movies)

    # Display the DataFrame as a table
    print(df)

else:
    print("Error fetching data. Status code:", response.status_code)
```

```python
df.drop(columns=["backdrop_path","adult","original_language","id","poster_path","video"], inplace=True)
```

```python
import requests

url1 = "https://api.themoviedb.org/3/genre/movie/list?language=en"

response = requests.get(url1, headers=headers)

print(response.text)

genres_data = response.json()
genre_map = {genre["id"]: genre["name"] for genre in genres_data.get("genres", [])}

# Replace genre_ids with genre names
df["genre_names"] = df["genre_ids"].apply(lambda ids: [genre_map[id] for id in ids])

# Drop the original genre_ids column if needed
df.drop(columns=["genre_ids"], inplace=True)

# Display the DataFrame with genre names
print(df)
```
